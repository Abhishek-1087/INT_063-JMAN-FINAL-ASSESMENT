```python
from datetime import timedelta
fake_data_list = []

# Define the list of possible training names
training_names = [
    "Introduction to Python Programming",
    "Web Development Fundamentals",
    "Data Science Essentials",
    "Machine Learning Basics",
    "Introduction to JavaScript",
    "Frontend Development with React",
    "Backend Development with Node.js",
    "Database Management with MongoDB",
    "Cloud Computing Fundamentals",
    "Cybersecurity Essentials",
]

# Generate fake data
for i in range(1, 100):
    certificates = random.sample(training_names, 4)
    start_date = fake.date_this_decade()
    end_date = start_date + timedelta(days=random.randint(5, 10))  # Random end date 5 to 10 days ahead
    training_data = {
        "trainingId": "Training_" + str(i),
        "trainingName": fake.random_element(elements=certificates),
        "trainingDescription": fake.sentence(),
        "trainerId": "Emp_" + str(i),
        "trainerName": fake.name(),
        "startDate": start_date,
        "endDate": end_date,
        "optimizedDuration": fake.random_int(min=1, max=40),
        "createdAt": fake.date_time_this_year(),
    }
    fake_data_list.append(training_data)

# Write data to CSV file
csv_file_path = "training_data.csv"
with open(csv_file_path, "w", newline="", encoding="utf-8") as csv_file:
    fieldnames = fake_data_list[0].keys()
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

    writer.writeheader()
    for training_data in fake_data_list:
        writer.writerow(training_data)

print(f"Training data has been saved to {csv_file_path}")
```

```python
import csv
import random

# Read data from CSV file
def read_employee_skills(filename):
    employees = {}
    with open(filename, 'r') as file:
        csv_reader = csv.DictReader(file)
        for row in csv_reader:
            emp_id = row['userId']
            skill = row['skills']
            if emp_id in employees:
                employees[emp_id].append(skill)
            else:
                employees[emp_id] = [skill]
    return employees

# Define projects
projects = [
    "Smart Home Automation System",
    "E-commerce Platform Redesign",
    "Personal Finance Tracker App",
    "AI-powered Virtual Assistant",
    "Online Learning Management System",
    "Autonomous Drone Delivery Service",
    "Healthcare Data Analytics Platform",
    "Blockchain-based Supply Chain Management",
    "Augmented Reality Shopping Experience",
    "Renewable Energy Management System",
    "Urban Traffic Management Solution",
    "Food Delivery Optimization Algorithm",
    "Mental Health Chatbot",
    "Real Estate Property Finder App",
    "Cybersecurity Incident Response Tool",
    "Smart City Waste Management System",
    "Online Event Ticketing Platform",
    "Agricultural Crop Monitoring System",
    "AI-driven Customer Service Chatbot",
    "Remote Team Collaboration Platform",
    "Online Medical Consultation App",
    "Language Translation Service",
    "Fitness and Nutrition Tracking App",
    "Virtual Reality Training Simulation",
    "Wildlife Conservation Monitoring System",
    "IoT-based Smart Agriculture Solution",
    "Emergency Response Management App",
    "Personalized News Aggregator",
    "Climate Change Prediction Model",
    "Smart Parking Management System"
]

# Function to assign projects to employees
def assign_projects(employees, skills):
    assignments = []
    for emp, skills_list in employees.items():
        # Shuffle the projects list to randomize the selection
        random.shuffle(projects)
        # Randomly select the number of projects (2 to 4)
        num_projects = random.randint(2, 4)
        # Ensure at least one project contains one of the employee's skills
        num_projects_with_skill = 1
        if len(skills_list) > 0:
            num_projects_with_skill += random.randint(0, min(len(skills_list), num_projects - 1))
        # Assign projects based on random selection
        assigned_projects = random.sample(projects, num_projects_with_skill)
        assigned_projects += random.sample([proj for proj in projects if proj not in assigned_projects], num_projects - num_projects_with_skill)
        # Add assignments to the list
        for proj in assigned_projects:
            # Randomly assign a skill from the employee's skills list to each project if available
            if skills_list:
                skill_used = random.choice(skills_list)
            else:
                skill_used = random.choice(skills)
            # Randomly assign acceptance or rejection for each project
            accepted = random.choice(['Accepted', 'Rejected' , 'Ongoing'])
            assignments.append({'Empid': emp, 'projectName': proj, 'majorSkill': skill_used, 'Status': accepted})
    return assignments

# Read employee skills from CSV file
employee_skills = read_employee_skills('fake_data.csv')

# List of skills
skills = [
    "Python", "JavaScript", "HTML", "CSS", "React", "Angular", "Vue.js", "Node.js", "Django", "Flask",
    "Java", "Kotlin", "Swift", "Objective-C", "C#", ".NET", "PHP", "Ruby", "SQL", "NoSQL",
    "MongoDB", "MySQL", "PostgreSQL", "Git", "Docker", "Kubernetes", "AWS", "Azure", "Google Cloud Platform",
    "RESTful API", "GraphQL", "CI/CD", "Agile Methodologies", "Scrum", "Test-Driven Development",
    "Behavior-Driven Development", "DevOps", "Continuous Integration", "Continuous Deployment", "Microservices",
    "Containerization", "Machine Learning", "Artificial Intelligence", "Data Science", "Data Analysis",
    "Data Visualization", "Statistical Analysis", "Cybersecurity", "Network Security", "Penetration Testing",
    "Cryptography", "Information Security", "UI/UX Design", "Wireframing", "Prototyping", "User Research",
    "Interaction Design", "Responsive Design", "Accessibility", "Product Management", "Agile Product Management",
    "Requirements Analysis", "User Stories", "Business Analysis", "Technical Writing", "Documentation",
    "Support", "Troubleshooting", "Problem Solving", "Communication", "Collaboration", "Leadership"
]

# Generate assignments
assignments = assign_projects(employee_skills, skills)

# Write assignments to a new CSV file
with open('employee_project_assignments.csv', 'w', newline='') as file:
    fieldnames = ['Empid', 'projectName', 'majorSkill', 'Status']
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(assignments)

print("CSV file created successfully.")
```
